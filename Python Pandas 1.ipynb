```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.shape
```

```python
### Filtering
booleans = []
for l in movies.duration:
    if l >= 200:
        booleans.append(True)
    else :
        booleans.append(False)
```

```python
booleans[0:10]
```

is_long= pd.Series(booleans)
is_long = movies.duration >= 200


```python
len(is_long)
```

```python
is_long.head()
```

```python
movies[is_long]
```

```python
### Alternatively, this is the actual method to apply filters
movies[movies.duration >= 200]
```

```python
###Selecting only the genre column for filters

movies[movies.duration >= 200].genre
```

```python
### Applying multiple filters

movies[(movies.duration >= 200) & (movies.genre=='Drama')] #and 
movies[(movies.duration >= 200) | (movies.genre=='Drama')] #or
```

###Selecting multiple genres

movies[movies.genre.isin(['Crime', 'Drama', 'Action'])]
